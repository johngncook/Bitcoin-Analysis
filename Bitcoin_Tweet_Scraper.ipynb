```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import re

# Function to Prepare and Clean Data for input into DataFrame Using Regular Expressions.
def parse_strlist(sl):
    clean = re.sub("[\[\],\s]","",sl)
    splitted = re.split("[\'\"]",clean)
    values_only = [s for s in splitted if s != '']
    return values_only

# Using Beautiful Soup to scrape the date and number of tweets from web page.
# Get URL and beautiful soup instance.
url = 'https://bitinfocharts.com/comparison/bitcoin-tweets.html'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# From scripts turn the script[4] into a string value to rip date and tweets out of the html.
scripts = soup.find_all('script')
StrList = str(scripts[4])
StrList = '[[' + StrList.split('[[')[-1]
StrList = StrList.split(']]')[0] +']]'
StrList = StrList.replace("new Date(", '').replace(')','')

# Then apply the parse_strlist function.
dataList = parse_strlist(StrList)

# Create lists to store dates and tweets.
date = []
tweet = []

# Loop through and append values to the lists.
for each in dataList:
    if (dataList.index(each) % 2) == 0:
        date.append(each)
    else:
        tweet.append(each)
# Put the lists to a DataFrame.
df = pd.DataFrame(list(zip(date, tweet)), columns=["Date","Tweets"])
```
